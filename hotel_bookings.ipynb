```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
#loading the dataset
df = pd.read_csv('hotel_bookings.csv')
df.head()
```

```python
print("Dataset shape:", df.shape)
print("\nColumn names:", df.columns.tolist())
print("\nDataset info:")
print(df.info())
df.describe()
```

```python
#Check missing values
missing = df.isnull().sum()
print(missing[missing > 0])
```

```python
#Handle missing values
df['children'] = df['children'].fillna(0)
df = df.drop(columns=['company'])
df['country'] = df['country'].fillna('Unknown')
df['agent'] = df['agent'].fillna(0)
df.isnull().sum().sum()
```

```python
#Fix data types
df['agent'] = df['agent'].astype(int)
df['children'] = df['children'].astype(int)
print(df.dtypes.head())
```

```python
# Exploratory Data Analysis

plt.figure(figsize=(6,4))
sns.countplot(x='is_canceled', data=df)
plt.title('Booking Cancellations (0 = Not Canceled, 1 = Canceled)')
plt.show()


plt.figure(figsize=(6,4))
sns.countplot(x='hotel', data=df)
plt.title('Hotel Types')
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x='hotel', y='adr', data=df)
plt.title('ADR (Average Daily Rate) by Hotel')
plt.show()
```

```python
#  Preprocessing
from sklearn.preprocessing import LabelEncoder, StandardScaler
cat_cols = ['hotel','meal','country','market_segment','distribution_channel','reserved_room_type','deposit_type','customer_type','reservation_status']

encoder = LabelEncoder()
for col in cat_cols:
    df[col] = encoder.fit_transform(df[col])


scaler = StandardScaler()
numeric_cols = ['lead_time','adr','stays_in_weekend_nights','stays_in_week_nights']
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

print("Preprocessing done!")
df.head()
```
